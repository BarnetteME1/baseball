# Is there a correlation between a players height and their batting average

```python
#Import the appropriate libraries
import pandas as pd
import matplotlib.pyplot as plt

#Do this to see more columns just to review what they initially look like
pd.set_option('display.max_columns', 50)

#This allows you to see the graphs in the notebook
%matplotlib inline
```

```python
#Loading the files that will be used.
batting = pd.read_csv('baseball_csv_files/batting.csv')
player_info = pd.read_csv('baseball_csv_files/player.csv')
```

```python
#Merge the files together so that I can use the batting info for batting average and the player info for height
batting_player_info = pd.merge(batting, player_info, on='player_id')
```

```python
# Taking a look at the information to figure out where about I should cut off for the at bat
batting_player_info.ab.describe()
```

```python
#Take out the people who were at bat less than 57 times as they may skew the data
batting_player_info = batting_player_info[batting_player_info.ab >= 57]
```

```python
#Looking at the info again to see how much is left and how much things have changed.
batting_player_info.ab.describe()
```

```python
#Creating a batting average by dividing the amount of hits by the amount of times at bat
batting_player_info['BatAvg'] = batting_player_info.h/batting_player_info.ab
```

```python
batting_player_info.plot(kind='scatter', x = 'BatAvg', y = 'height', figsize = (20, 10))
```

### It seems like there might be a bit of a difference. If you look at those below around 68-70 they do have a higher low battng average, but some of but besides 64 they seem to not have as high a batting average either. When you get a bit taller 70-74 they seem to have a higher low batting average amount, but also have a better high batting average, with one at 74 being out around .5. When they get around 76-80 It looks like the low end goes up again, but again the high end starts to go down again.
